# **4. Seasonality & Patterns:**

Does bitcoin have seasonal patterns such as trends for different months?

Are there cyclical patterns in bitcoin price?

```python
# Convert 'date' to datetime
df['date'] = pd.to_datetime(df['date'])

# Set as index
df.set_index('date', inplace=True)

print(df.info())  # Check data types
print(df.head())  # Preview dataset
```

```python
# Extract month from index
df['Month'] = df.index.month
```

```python
# Group by Month and calculate average BTC price
monthly_avg = df.groupby('Month')['btc_price'].mean()
```

```python
# Plot Monthly Average Prices
plt.figure(figsize=(12, 6))
sns.barplot(x=monthly_avg.index, y=monthly_avg.values, palette="coolwarm")
plt.xticks(range(12), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.title("Bitcoin Average Price by Month")
plt.xlabel("Month")
plt.ylabel("Average Bitcoin Price (USD)")
plt.show()
```

```python
# Ensure 'date' is the index
df = df.sort_index()

# Resample to Monthly Data and Compute % Change
df_monthly = df[['btc_price', 'global_m2']].resample('M').mean()  # Resample to monthly average
df_monthly['btc_price_change'] = df_monthly['btc_price'].pct_change() * 100  # Convert to %
df_monthly['global_m2_change'] = df_monthly['global_m2'].pct_change() * 100  # Convert to %

# Plot Comparison
fig, ax1 = plt.subplots(figsize=(12, 6))

# Bitcoin Price Change
ax1.set_xlabel("Date")
ax1.set_ylabel("Bitcoin Monthly % Change", color='blue')
ax1.plot(df_monthly.index, df_monthly['btc_price_change'], color='blue', label="Bitcoin Price Change", linewidth=2)
ax1.tick_params(axis='y', labelcolor='blue')

# Create second axis for Global M2
ax2 = ax1.twinx()
ax2.set_ylabel("Global M2 Monthly % Change", color='red')
ax2.plot(df_monthly.index, df_monthly['global_m2_change'], color='red', linestyle='dashed', label="Global M2 Change", linewidth=2)
ax2.tick_params(axis='y', labelcolor='red')

# Title & Legend
fig.suptitle("Bitcoin Monthly Change vs. Global M2 Monthly Change")
ax1.legend(loc='upper left')
ax2.legend(loc='upper right')

plt.show()
```

```python
# Ensure 'date' is the index
df = df.sort_index()

# Resample to Monthly Data and Compute % Change
df_monthly = df[['btc_price', 'global_m2']].resample('M').mean()

# Compute Monthly Percentage Changes
df_monthly['btc_price_change'] = df_monthly['btc_price'].pct_change() * 100  # Convert to %
df_monthly['global_m2_change'] = df_monthly['global_m2'].pct_change() * 100  # Convert to %

# Shift Global M2 forward by 108 days (~3.6 months)
df_monthly['global_m2_change_shifted'] = df_monthly['global_m2_change'].shift(-3)

# Plot Comparison
fig, ax1 = plt.subplots(figsize=(12, 6))

# Bitcoin Price Change
ax1.set_xlabel("Date")
ax1.set_ylabel("Bitcoin Monthly % Change", color='blue')
ax1.plot(df_monthly.index, df_monthly['btc_price_change'], color='blue', label="Bitcoin Price Change", linewidth=2)
ax1.tick_params(axis='y', labelcolor='blue')

# Create second axis for Shifted Global M2
ax2 = ax1.twinx()
ax2.set_ylabel("Global M2 Monthly % Change (Shifted 108 Days)", color='red')
ax2.plot(df_monthly.index, df_monthly['global_m2_change_shifted'], color='red', linestyle='dashed', label="Global M2 Change (Shifted)", linewidth=2)
ax2.tick_params(axis='y', labelcolor='red')

# Title & Legend
fig.suptitle("Bitcoin Monthly Change vs. Global M2 Change (Shifted 108 Days)")
ax1.legend(loc='upper left')
ax2.legend(loc='upper right')

plt.show()
```

# **5. Outliers & Anomalies:**

Are there outliers in bitcoin price? What external factors caused them?


```python
# Create a boxplot for Bitcoin price
plt.figure(figsize=(10, 5))
sns.boxplot(x=df['btc_price'], color='blue')

# Title
plt.title("Boxplot of Bitcoin Prices")
plt.xlabel("Bitcoin Price (USD)")

plt.show()
```

```python
# Compute Q1 (25th percentile) and Q3 (75th percentile)
Q1 = df['btc_price'].quantile(0.25)
Q3 = df['btc_price'].quantile(0.75)

# Compute Interquartile Range (IQR)
IQR = Q3 - Q1

# Define bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Find outliers
outliers = df[(df['btc_price'] < lower_bound) | (df['btc_price'] > upper_bound)]

print(f"Number of outliers: {outliers.shape[0]}")
print(outliers[['btc_price']].head())  # Show some outliers
```

```python
# Calculate Rolling Mean & Standard Deviation
df['rolling_mean'] = df['btc_price'].rolling(window=30).mean()
df['rolling_std'] = df['btc_price'].rolling(window=30).std()

# Define outlier threshold (mean Â± 2.5 std deviations)
df['outlier'] = (df['btc_price'] > df['rolling_mean'] + 2.5 * df['rolling_std']) | \
                (df['btc_price'] < df['rolling_mean'] - 2.5 * df['rolling_std'])

# Plot Bitcoin Price with Outliers
plt.figure(figsize=(12, 6))
plt.plot(df.index, df['btc_price'], label="Bitcoin Price", color='blue', alpha=0.7)
plt.scatter(df[df['outlier']].index, df[df['outlier']]['btc_price'], color='red', label="Outliers", marker='o')
plt.title("Bitcoin Price with Detected Outliers")
plt.xlabel("Date")
plt.ylabel("Bitcoin Price (USD)")
plt.legend()
plt.show()
```

# **6. Time-Series Analysis:**

How do Bitcoin prices respond to changes in M2 money supply over time? Is there a lag between an economic indicator and a price change?

```python
# Compute Monthly Percentage Change
df['btc_price_change'] = df['btc_price'].pct_change() * 100
df['global_m2_change'] = df['global_m2'].pct_change() * 100
```

```python
# computer correlation between m2 and btc price changes
corr = df[['btc_price_change', 'global_m2_change']].corr()

# Heatmap of correlation
plt.figure(figsize=(6, 4))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title("Correlation Between Bitcoin Price Change and Global M2 Change")
plt.show()
```

```python
# Check lagged correlation using the cross correlation function
from statsmodels.tsa.stattools import acf
from statsmodels.tsa.stattools import ccf
```

```python
# Compute cross-correlation between M2 and Bitcoin price
lags = range(-12, 13)  # Check lags from -12 to +12 months
ccf_values = [df['global_m2_change'].shift(lag).corr(df['btc_price_change']) for lag in lags]

# Plot cross-correlation
plt.figure(figsize=(10, 5))
plt.bar(lags, ccf_values, color='purple', alpha=0.7)
plt.axhline(0, linestyle='--', color='black')
plt.xlabel("Lag (Months)")
plt.ylabel("Cross-Correlation")
plt.title("Lagged Correlation Between Bitcoin Price & Global M2")
plt.show()
```
