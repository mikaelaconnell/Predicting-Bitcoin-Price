```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# import csv file
df = pd.read_csv("final_merged_data.csv")
```

```python
df.info()
```

```python
# rename m2_supply column to us_m2
df.rename(columns={'m2_supply': 'us_m2'}, inplace=True)
```

```python
df.info()
```

```python
df.head()
```

```python
# change us m2 to be the same scale as china, japan, and europe
df["us_m2_scaled"] = df["us_m2"] * 1000000000
```

```python
df[["us_m2_scaled", "japan_m2", "china_m2", "europe_m2", "brazil_m2", "india_m2", "canada_m2", "russia_m2"]].describe()
```

```python
# Convert to USD

exchange_rates = {
    "JPY": 150, "CNY": 7, "EUR": 1.1, "BRL": 5, "INR": 83, "CAD": 1.3, "RUB": 95
}

df["japan_m2_usd"] = df["japan_m2"] * exchange_rates["JPY"]
df["china_m2_usd"] = df["china_m2"] * exchange_rates["CNY"]
df["europe_m2_usd"] = df["europe_m2"] * exchange_rates["EUR"]
df["brazil_m2_usd"] = df["brazil_m2"] * exchange_rates["BRL"]
df["india_m2_usd"] = df["india_m2"] * exchange_rates["INR"]
df["canada_m2_usd"] = df["canada_m2"] * exchange_rates["CAD"]
df["russia_m2_usd"] = df["russia_m2"] * exchange_rates["RUB"]
```

```python
df.info()
```

```python
# plot us m2 scales, japan m2 usd, china m2 usd, and europe me usd
plt.figure(figsize=(12, 6))
plt.plot(df['date'], df[['us_m2_scaled', 'japan_m2_usd', 'china_m2_usd', 'europe_m2_usd']], label=["US M2", "Japan M2", "China M2", "Europe M2"])
plt.xlabel("Date")
plt.ylabel("Value")
plt.title("M2 Money Supply")
plt.legend()
plt.show()
```

```python
# Plot US M2
plt.figure(figsize=(12, 6))
plt.plot(df['date'], df[['us_m2']], label=["US M2"], linestyle='dashed')

plt.xlabel("Date")
plt.ylabel("Value (Billions)")
plt.title("US M2 Money Supply")
plt.legend()
plt.show()
```

```python
# plot Us m2 scaled
plt.figure(figsize=(12, 6))
plt.plot(df['date'], df[['us_m2_scaled']], label=["US M2 Scaled"], linestyle='dashed')
plt.xlabel("Date")
plt.ylabel("Value")
plt.title("US M2 Money Supply")
plt.legend()
plt.show()
```

```python
#plot japan m2
plt.figure(figsize=(12, 6))
plt.plot(df['date'], df[['japan_m2']], label=["Japan M2"], linestyle='dashed')
plt.xlabel("Date")
plt.ylabel("Value")
plt.title("Japan M2 Money Supply")
plt.legend()
plt.show()
```

```python
# plot europe m2
plt.figure(figsize=(12, 6))
plt.plot(df['date'], df[['europe_m2']], label=["Europe M2"], linestyle='dashed')
plt.xlabel("Date")
plt.ylabel("Value")
plt.title("Europe M2 Money Supply")
plt.legend()
plt.show()
```

```python
# Create new column that represents the global M2
df["global_m2"] = df["us_m2_scaled"] + df["japan_m2_usd"] + df["china_m2_usd"] + df["europe_m2_usd"] + df["brazil_m2_usd"] + df["india_m2_usd"] + df["canada_m2_usd"] + df["russia_m2_usd"]
```

```python
df[["global_m2", "us_m2_scaled", "japan_m2_usd", "china_m2_usd", "europe_m2_usd", "brazil_m2_usd", "india_m2_usd", "canada_m2_usd", "russia_m2_usd"]].describe()
```

```python
# plot global m2
plt.figure(figsize=(12, 6))
plt.plot(df['date'], df[['global_m2']], label=["Global M2"], linestyle='dashed')
plt.xlabel("Date")
plt.ylabel("Value")
plt.title("Global M2 Money Supply")
plt.legend()
#plt.show()
```

```python
# How has bitcoin price changed overtime
plt.figure(figsize=(12, 6))
plt.plot(df['date'], df['btc_price'], label="Bitcoin Price")
plt.xlabel("Date")
plt.ylabel("Price (USD)")
plt.title("Bitcoin Price Over Time")
```
